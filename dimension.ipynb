```python
#rounding off starttime and stoptime
df['starttime'] = pd.to_datetime(df['starttime']).dt.floor('s')  #.floor() - resets the value to nearest zero, removing frational part of seconds
df['stoptime'] = pd.to_datetime(df['stoptime']).dt.floor('s')
```

```python
#making a tripduration column. 
df['tripduration'] = (df['stoptime'] - df['starttime']).dt.total_seconds().astype(int)
```
